```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("transformed.csv")
df.head()
```

```python
df.columns
```

## Brand products count

```python
brands_count = df['Brand'].value_counts()
plt.bar(brands_count.index, brands_count.values)
plt.xlabel('Brand')
plt.ylabel('Count')
plt.title('Brand Product Counts')
plt.xticks(rotation=90)
plt.show()
```

## Price Distribution


```python
plt.hist(df['sale_price'], bins = 30, edgecolor='black')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Price Distribution')
plt.xticks(rotation=90)
plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))

axes[0, 0].scatter(df['sale_price'], df["average_rating"])
axes[0, 0].set_xlabel('Price')
axes[0, 0].set_ylabel('Average Rating')
axes[0, 0].set_title('Price and rating')

axes[0, 1].scatter(df["Model Year"].fillna("2023"), df['sale_price'])
axes[0, 1].set_xlabel('Built year')
axes[0, 1].set_ylabel('Price')
axes[0, 1].tick_params(axis='x', labelrotation=45)
axes[0, 1].set_title('Built year and price')

axes[1, 0].scatter(df["Battery Capacity"].fillna("Not Specified"), df['sale_price'])
axes[1, 0].set_xlabel('Battery Capacity')
axes[1, 0].set_xlabel('Price')
axes[1, 0].tick_params(axis='x', labelrotation=45)
axes[1, 0].set_title('Battery Capacity and Price')

axes[1, 1].scatter(df['Screen Size (inches)'].fillna("Not Specified"), df["sale_price"])
axes[1, 1].set_xlabel('Screen Size')
axes[1, 1].set_xlabel('Price')
axes[1, 1].tick_params(axis='x', labelrotation=45)
axes[1, 1].set_title('Screen Size and Price')
```
